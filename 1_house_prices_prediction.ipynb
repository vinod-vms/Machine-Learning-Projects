```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# we are now loading the dataset

X, y = load_boston(return_X_y=True)
```

```python
#importing the model from sklearn (actually this is a Python class)
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
regressor2 = LinearRegression()
```

```python
# This is the model. I'm calling it regressor
regressor = KNeighborsRegressor()
```

```python
#Spilting the dataset
import sklearn.model_selection as model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split( X, y, test_size = 1/4, random_state = 0) 
```

```python
#Training the model. In ML parlance this is called 'fitting'
regressor.fit(X_train, y_train)
regressor2.fit(X_train, y_train)
```
