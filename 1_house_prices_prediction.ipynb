```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
#loading the dataset

X, y = load_boston(return_X_y=True)
```

```python
#importing the models from sklearn (actually these are Python classes)
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
```

```python
# These are the models. I'm calling it regressor
regressor = KNeighborsRegressor()
regressor2 = LinearRegression()
```

```python
#Spilting the dataset
import sklearn.model_selection as model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split( X, y, test_size = 1/4, random_state = 0) 
```

```python
#Training the models. In ML parlance this is called 'fitting'
regressor.fit(X_train, y_train)
regressor2.fit(X_train, y_train)
```

```python
#Visualizing the testing aka the actual prediction
plt.scatter(y_test, regressor.predict(X_test), color = 'red');
plt.title("Actual_vs_Predicted")
plt.xlabel("Actual Prices (in 000 of dollars)")
plt.ylabel("Predicted Prices (in 000 of dollars)")
plt.grid(True)
plt.axis([0,55, 0,55])
```

```python
plt.scatter(y_test, regressor2.predict(X_test), color = 'red');
plt.title("Actual_vs_Predicted")
plt.xlabel("Actual Prices (in 000 of dollars)")
plt.ylabel("Predicted Prices (in 000 of dollars)")
plt.grid(True)
plt.axis([0,55, 0,55])
```
